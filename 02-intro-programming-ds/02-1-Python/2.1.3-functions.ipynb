# Funciones con listas

```python
def greet(names):
    for name in names:
        msg = "Hello, " + name.title() + "!"
        print(msg)
        
```

```python
usernames = ['margot', 'tina', 'alex']
greet(usernames)
```

```python
def print_names(names):
    names_copy = list(names)
    for i in range(len(names_copy)):
        name = names_copy.pop()
        print(name)
```

```python
def print_names(names):
    for i in range(len(names)):
        name = names.pop()
        print(name)
```

# Paso arbitrario de número de argumentos


```python
def make_pizza(*toppings, num=3):
    for _ in range(num):
        print(toppings)
```

```python
make_pizza('pepperoni', num=1)
```

```python
make_pizza('queso', 'jamón', 'piña')
```

```python
def make_pizza(*toppings):
    for topping in toppings:
        print("- " + topping)
```

```python
make_pizza('queso', 'jamón', 'piña')
```

```python
def make_pizza(size, *toppings):
    print("Pizza de tamaño" + str(size))
    for topping in toppings:
        print("- " + topping)
```

```python
make_pizza('queso', 'jamón', 'piña')
```

```python
def profile(first, last, **user_info):
    profile = {}
    profile['fist_name'] = first,
    profile['last_name'] = last,
    for key, value in user_info.items():
        profile[key] = value
    return profile
```

```python
user_profile = profile('ana', 'millan', location='oaxaca', title='miss')
```
