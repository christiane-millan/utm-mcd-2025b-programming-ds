# Pandas

## Series

```python
import pandas as pd
import numpy as np
```

```python
students = pd.Series(['Robert', 'Charly', 'George', 'Leo'])
students
```

```python
students = pd.Series(['Robert', 'Charly', 'George', 'Leo'], index=[1, 30, 10, 7])
students
```

```python
dict = {1:'Robert', 7:'Charly', 10:'George', 30:'Leo'}
```

```python
students = pd.Series(dict)
students
```

```python
my_array = np.array([10, 20, 30, 40, 50])
my_series = pd.Series(my_array, index = ["num1", "num2", "num3", "num4", "num5"])
my_series
```

## Acceso a series

```python
my_series[2]
```

```python
my_series['num3']
```

```python
my_series = pd.Series([5, 8, 2, 11, 9])
```

## DataFrames

```python
scores = [['Mathematics', 85], ['English', 91], ['History', 95]]
```

```python
my_df = pd.DataFrame(scores, columns=['Subject', 'Score'])
my_df
```

```python
scores = {
    'Subject':["Mathematics", "History", "English", "Science", "Arts"],
    'Score':[98, 75, 68, 82, 99]
}
```

```python
my_df = pd.DataFrame(scores)
my_df
```

## Operaciones básicas `head` y `tail`

```python
my_df.head(2)
```

```python
my_df.tail(3)
```

## Resumen de un dataframe

```python
my_df.info()
```

## Estadísiticos básicos

```python
my_df.describe()
```
