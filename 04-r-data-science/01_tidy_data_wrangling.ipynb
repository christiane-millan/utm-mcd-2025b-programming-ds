# verbo arrange

```r
gapminder %>% arrange(lifeExp)
```

```r
gapminder %>% arrange(desc(lifeExp))
```

```r
gapminder %>% filter(year == 1957) %>% arrange(desc(pop))
```

# verbo mutate

```r
gapminder %>% mutate(lifeExp = 12 * lifeExp)
```

```r
gapminder %>% mutate(lifeExpMonths = 12 * lifeExp)
```

```r
gapminder %>% 
filter(year == 2007) %>% 
mutate(lifeExpMonths = 12 * lifeExp) %>% 
arrange(desc(lifeExpMonths))
```

```r
library(ggplot2)
```

```r
gapminder_1957 <- gapminder %>% filter(year == 1957)

head(gapminder_1957)
```

```r
ggplot(gapminder_1957, aes(x = gdpPercap, y = lifeExp)) + geom_point()
```

```r
ggplot(gapminder_1957, aes(x=pop, y=gdpPercap)) + geom_point() + scale_x_log10() + scale_y_log10()
```

```r
ggplot(gapminder_1957, aes(x=pop, y=lifeExp, color=continent, size=gdpPercap)) + geom_point() + scale_x_log10() 
```

# verbo summarize

```r
gapminder %>% summarize(medianLifeExp = median(lifeExp))
```

```r
gapminder %>% filter(year == 1957) %>% 
summarize(medianLifeExp = median(lifeExp))
```

```r
gapminder %>% filter(year == 1957) %>% summarize(medianLifeExp = median(lifeExp), maxGdpPercap = max(gdpPercap))
```

# Verbo group by

```r
gapminder %>% group_by(year) %>% 
summarize(medianLifeExp = median(lifeExp), maxGdpPercap = max(gdpPercap))
```

```r
gapminder %>% group_by(continent, year) %>% 
summarize(medianLifeExp = median(lifeExp), maxGdpPercap = max(gdpPercap))
```

```r
by_year <- gapminder %>% 
group_by(year) %>% 
summarize(medianLifeExp = median(lifeExp),
maxGdpPercap = max(gdpPercap))

ggplot(by_year, aes(x=year, y=medianLifeExp)) + geom_point() 
```

```r
gapminder
```

```r
by_year_continent <- gapminder %>% group_by(year, continent) %>% 
summarize(medianGdpPercap = median(gdpPercap))
```

```r
ggplot(by_year_continent, aes(x=year, y=medianGdpPercap, color=continent)) + geom_point()
```

```r
by_continent_2007 <- gapminder %>% filter(year == 2007) %>% group_by(continent) %>% summarize(medianLifeExp = median(lifeExp), medianGdpPercap = median(gdpPercap))
```

```r
ggplot(by_continent_2007, aes(x=medianGdpPercap, y=medianLifeExp, color=continent)) + geom_point()
```
